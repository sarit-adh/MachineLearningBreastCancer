```python
from sklearn import tree
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.cross_validation import cross_val_score
from sklearn.tree import export_graphviz
import pydotplus
from subprocess import call
```

```python
file = pd.read_csv("data.csv")
df = pd.DataFrame(file)
df.head()
```

```python
df.tail()
```

```python
df.drop(df.columns[32],axis=1, inplace=True) #remove the empty column
```

```python
df.tail()
```

```python
diagnosis_type = df['diagnosis'].unique()
diagnosis_type
```

```python
#We need to encode the diagnosis to integer
def encoding_target(df, D_type):
    df_mod = df.copy()
    convert_int = {name: n for n, name in enumerate(D_type)}
    df_mod["D_type"]= df_mod["diagnosis"].replace(convert_int)
    
    return(df_mod)
    
```

```python
new_df = encoding_target(df, diagnosis_type)
new_df[["diagnosis", "D_type"]].head()
```

```python
new_df[["diagnosis", "D_type"]].tail()
```

```python
features = list(new_df.columns[2:32])
features
```

```python
y = new_df["D_type"]
X= new_df[features]
dt = tree.DecisionTreeClassifier(criterion='gini')
dt.fit(X,y)
```

```python
export_graphviz(dt,out_file='dt.dot', feature_names=features, class_names=None, filled=True, rounded=True, special_characters=True )
```
